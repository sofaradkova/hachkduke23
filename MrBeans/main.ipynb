```python
evilBook = open("./art.txt", "r")
evil = evilBook.read()
evilBook.close()
```

```python
from transformers import pipeline

summerizer = pipeline("summarization", max_length=100)
output = summerizer(evil)
KindBook = open("./out.txt", "w")
KindBook.write(output[0]['summary_text'])
KindBook.close()
```
